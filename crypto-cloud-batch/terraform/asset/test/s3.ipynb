```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("S3Test") \
    .master("local[*]") \
    .config("spark.jars", ",".join([
        "../jars/hadoop-aws-3.3.4.jar",
        "../jars/aws-java-sdk-bundle-1.12.262.jar"
    ])) \
    .getOrCreate()
```

```python
data = [(1, "Alice"), (2, "Bob")]
df = spark.createDataFrame(data, ["id", "name"])
```

```python
df.show()
```

```python
df.write.mode("overwrite").parquet("s3a://pyspark-local-test-bucket/testdata/")
print("✅ Successfully wrote parquet to S3!")
```
